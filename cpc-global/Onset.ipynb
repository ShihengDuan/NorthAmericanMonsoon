```python
import xarray as xa
import numpy as np
from matplotlib import pyplot as plt
```

```python
def is_monsoon_precip(month):
    return (month <= 9) & (month >= 6)

data = xa.open_dataarray('/tempest/duan0000/exprecip/cpc-global/NAM_sub_precip')
monsoon_precip = data.sel(time=(is_monsoon_precip(data.time.dt.month)))
monsoon_precip = monsoon_precip.sel(time=(monsoon_precip.time.dt.year < 2019))
```

```python
monsoon_precip
```

```python
days = []
sub_id = 1
sub_precip = monsoon_precip.sel(sub_id=sub_id)
for year in range(1979, 2019):
    year_precip = sub_precip.sel(time=(sub_precip.time.dt.year==year))
    for i in range(122-4):
        if year_precip.data[i]>=1 and year_precip.data[i+1]>=1 and year_precip.data[i+2]>=1:
            # print(year, i)
            days.append(i)
            break
print(len(days))
print(np.median(days), np.quantile(days, .75)-np.quantile(days, .25), np.min(days), np.max(days))
plt.boxplot(days)
plt.show()
```

# Consecutive 5-days

```python
initial = np.datetime64('1979-06-01')
for sub_id in range(1, 8):
    days = []
    sub_precip = monsoon_precip.sel(sub_id=sub_id)
    for year in range(1979, 2019):
        year_precip = sub_precip.sel(time=(sub_precip.time.dt.year==year))
        for i in range(122-4):
            if year_precip.data[i]>=1 and year_precip.data[i+1]>=1 and year_precip.data[i+2]>=1 and\
            year_precip.data[i+3]>=1 and year_precip.data[i+4]>=1:
                # print(year, i)
                days.append(i)
                break
    # print(len(days))
    print(sub_id)
    delta = np.median(days)
    print(len(days), delta, ' ', initial + np.timedelta64(int(delta), 'D'))
    # print(np.median(days), np.quantile(days, .75)-np.quantile(days, .25), np.min(days), np.max(days))
    plt.boxplot(days, positions=[sub_id])
plt.show()
```

# Max precip

```python
initial = np.datetime64('1979-06-01')
for sub_id in range(1, 8):
    print(sub_id)
    days = []
    sub_precip = monsoon_precip.sel(sub_id=sub_id)
    for year in range(1979, 2019):
        year_precip = sub_precip.sel(time=(sub_precip.time.dt.year==year))
        days.append(np.max(year_precip))
    # print(len(days))
    delta = np.median(days)
    print(len(days), delta)
    # print(np.median(days), np.quantile(days, .75)-np.quantile(days, .25), np.min(days), np.max(days))
    plt.boxplot(np.array(days).flatten(), positions=[sub_id])
plt.show()
```

# Max precip date

```python
initial = np.datetime64('1979-06-01')
for sub_id in range(1, 8):
    print(sub_id)
    days = []
    sub_precip = monsoon_precip.sel(sub_id=sub_id)
    for year in range(1979, 2019):
        year_precip = sub_precip.sel(time=(sub_precip.time.dt.year==year))
        days.append(np.argmax(year_precip.data))
    # print(len(days))
    delta = np.median(days)
    print(len(days), delta, ' ', initial + np.timedelta64(int(delta), 'D'))
    # print(np.median(days), np.quantile(days, .75)-np.quantile(days, .25), np.min(days), np.max(days))
    plt.boxplot(np.array(days).flatten(), positions=[sub_id])
plt.show()
```

# Standard variance

```python
initial = np.datetime64('1979-06-01')
for sub_id in range(1, 8):
    print(sub_id)
    days = []
    sub_precip = monsoon_precip.sel(sub_id=sub_id)
    for year in range(1979, 2019):
        year_precip = sub_precip.sel(time=(sub_precip.time.dt.year==year))
        days.append(np.std(year_precip.data))
    # print(len(days))
    delta = np.median(days)
    print(len(days), delta)
    # print(np.median(days), np.quantile(days, .75)-np.quantile(days, .25), np.min(days), np.max(days))
    plt.boxplot(np.array(days).flatten(), positions=[sub_id])
plt.show()
```
